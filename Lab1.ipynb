```python
#Simple Linear Regression
import numpy as np

def mean(values):
  return sum(values) / float(len(values))

def variance(values, mean):
  return sum([(x - mean)**2 for x in values])

def covariance(x, mean_x, y, mean_y):
  covar = 0.0
  for i in range(len(x)):
    covar += (x[i] - mean_x) * (y[i] - mean_y)
  return covar

def linear_regression(x, y):
  mean_x = mean(x)
  mean_y = mean(y)
  b1 = covariance(x, mean_x, y, mean_y)/variance(x,mean_x)
  b0 = mean_y - b1 * mean_x
  return [b0, b1]

x = np.asarray([1,2,3,4,5])
y = np.asarray([1,3,5,7,9])
b0, b1 = linear_regression(x, y)
print("b0:", b0)
print("b1:", b1)
```

```python
# AND logic gate
import numpy as np

x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.asarray([0, 0, 0, 1])
y_pred = np.asarray([0, 0, 0, 0])
w = np.asarray([1, 1])
t=1.5

def step(_sum):
    if (_sum > 1.5):
        return 1
    else:
        return 0

for i in range(x.shape[0]):
    n= x[i]
    _sum = np.dot(n,w)
    y_pred[i] = step(_sum)

print(y_pred)

#Plotting the decision boundary
import matplotlib.pyplot as plt
slope=-w[0]/w[1]
intercept=(t)/w[1]
x_plane=np.linspace(-2,2,10)
y_plane= slope * x_plane + intercept
plt.plot(x_plane-0.1,y_plane-0.1,'-')
plt.scatter(x[:,0],x[:,1],c=y)
plt.xlim([-0.2,1.2])
plt.ylim([-0.2,1.2])
plt.show()
```

```python
#OR Logic Gate
import numpy as np

x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.asarray([0, 1, 1, 1])
y_pred = []
w = np.asarray([1, 1])
t=1

def step(_sum):
    if (_sum >= 1):
        return 1
    else:
        return 0

for i in range(x.shape[0]):
    n= x[i]
    _sum = np.dot(n,w)
    y_pred.append(step(_sum))

print(y_pred)

#Plotting the decision boundary
import matplotlib.pyplot as plt
slope=-w[0]/w[1]
intercept=(t)/w[1]
x_plane=np.linspace(-2,2,10)
y_plane= slope * x_plane + intercept
plt.plot(x_plane-0.1,y_plane-0.1,'-')
plt.scatter(x[:,0],x[:,1],c=y)
plt.xlim([-0.2,1.2])
plt.ylim([-0.2,1.2])
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt

X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.asarray([0,0,0,1])
h = .02

clf = Perceptron().fit(X,Y)

x_min, x_max = X[:, 0].min() - 1, X[:,0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:,1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
#Plot the decision boundary
fig, ax = plt.subplots()
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
#Put the result into a color plot
Z = Z.reshape(xx.shape)
ax.contourf(xx, yy, Z, cmap=plt.cm.gist_earth)
ax.axis('off')
ax.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.gist_earth)
ax.set_title('Perceptron')
```
